```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset

df = pd.read_csv('Copper_dataset.csv')
df.head()
```

```python
# Check for null values
null_values = df.isnull().sum()

# Display the result
print(null_values)
```

```python
df.shape
```

```python
# # Convert values starting with '00000' in 'Material_Reference' to NaN

#df.loc[df['material_ref'].str.startswith('00000'),'material_ref']=pd.NA

df['material_ref'] = np.where(df['material_ref'].str.startswith('00000'),np.nan,df['material_ref'])

print(df)

```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# check whether the column is continuous or categorical

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        print(f"{column} is a continuous column.")
    elif pd.api.types.is_object_dtype(df[column]):
        print(f"{column} is a categorical column.")
    else:
        print(f"{column} has another data type.")
```

### categorical column filling using mode

```python
# filling null values with mode
categorical_columns = ['country','application','status','customer']

for column in categorical_columns:
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)

```

```python
df.isnull().sum()
```

### continuous column filling using mean

```python
continuous_columns = ['thickness', 'selling_price']

for column in continuous_columns:
    mean_value = df[column].mean()
    df[column].fillna(mean_value,inplace=True )
```

```python
df.isnull().sum()
```

```python
df['item_date'].fillna(0, inplace=True)

df['delivery date'].fillna(0,inplace=True)

df['id'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['material_ref'].fillna('Not Specified', inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.drop('material_ref', axis=1)
```

```python
df.isnull().sum()
```

### Status column value replace

```python

# Replace 'WON' with 'Success' and 'LOST' with 'Failure'
df['status'] = df['status'].map({'Won': 'Success', 'Lost': 'Failure'})

# Remove rows with other values in the 'STATUS' column
df = df[df['status'].isin(['Success', 'Failure'])]

```

```python
df.to_csv('Copper_dataset_v4.csv', index=False, encoding='utf-8-sig')
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Convert 'quantity tons' and 'selling_price' columns to numeric, handling errors
df['quantity tons'] = pd.to_numeric(df['quantity tons'], errors='coerce')
df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')
```

#### histograms to visualize the distribution of numerical columns

```python
# Plot histograms to visualize the distribution of numerical columns
df.hist(bins=20, figsize=(15,10))
plt.show()
```

#### Outliers

```python
# Visualize boxplots to identify outliers in 'Selling_Price'

sns.boxplot(x=df['selling_price'])
plt.show()
```

```python
# Calculate IQR for 'Selling_Price'
Q1 = df['selling_price'].quantile(0.25)
Q3 = df['selling_price'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['selling_price'] < lower_bound) | (df['selling_price'] > upper_bound)]

# Print the outliers
print("Outliers:")
print(outliers)

# Visualize boxplot with outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['selling_price'])
plt.title('Boxplot of Selling_Price with Outliers')
plt.show()
```

```python
# Visualize violinplot with outliers

plt.figure(figsize=(8, 6))
sns.violinplot(x=df['selling_price'])
plt.title('ViolinPlot of Selling_Price with Outliers')
plt.show()
```

```python
# Visualize displot with outliers

plt.figure(figsize=(8, 6))
sns.displot(x=df['selling_price'])
plt.title('Displot of Selling_Price with Outliers')
plt.show()
```

#### Remove outliers

```python
# Remove outliers
df_no_outliers = df[(df['selling_price'] >= lower_bound) & (df['selling_price'] <= upper_bound)]

# Print the number of removed outliers
num_outliers_removed = len(df) - len(df_no_outliers)
print(f"Number of outliers removed: {num_outliers_removed}")

# Visualize boxplot without outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_no_outliers['selling_price'])
plt.title('Boxplot of Selling_Price without Outliers')
plt.show()
```

```python
# Visualize violinplot without outliers

plt.figure(figsize=(8, 6))
sns.violinplot(x=df_no_outliers['selling_price'])
plt.title('Violinplot of Selling_Price without Outliers')
plt.show()
```

```python
# Visualize displot without outliers

plt.figure(figsize=(8, 6))
sns.displot(x=df_no_outliers['selling_price'])
plt.title('Displot of Selling_Price without Outliers')
plt.show()
```

#### Skewness

```python
# Measure skewness of 'Selling_Price'
selling_price_skewness = df['selling_price'].skew()
print(f"Selling_Price Skewness: {selling_price_skewness}")
```

#### visualization before skewness 

```python
sns.set(style="whitegrid")  # Set the style for the plot

# Plot the histogram and KDE for 'Selling_Price'
plt.figure(figsize=(10, 6))
sns.histplot(df['selling_price'], kde=True, color='blue')

# Add labels and title
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.title('Distribution of Selling Price')

# Display skewness value on the plot
plt.text(0.9, 0.9, f'Skewness: {selling_price_skewness:.2f}', transform=plt.gca().transAxes, ha='right', va='top')

# Show the plot
plt.show()
```

#### Log transformation approach

```python
# Log transformation of 'Selling_Price'
df['log_selling_price'] = np.log1p(df['selling_price'])

# Plot histograms before and after log transformation
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.title('Original Selling_Price Distribution')
sns.histplot(df['selling_price'], bins=20, kde=True)

plt.subplot(1, 2, 2)
plt.title('Log-transformed Selling_Price Distribution')
sns.histplot(df['log_selling_price'], bins=20, kde=True)

plt.show()
```

#### After log transformation

```python
log_selling_price_skewness = df['log_selling_price'].skew()
print(f"Log-Transformed Selling_Price Skewness: {log_selling_price_skewness}")
```

#### Visualization after log transformation

```python
# histplot

sns.histplot(df['log_selling_price'], kde=True)
plt.title('Log-Transformed Selling_Price Distribution')
plt.xlabel('Log-Transformed Selling_Price')
plt.ylabel('Frequency')
plt.show()
```

#### kdeplot
